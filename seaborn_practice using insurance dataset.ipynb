```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./insurance.csv')
```

```python
df.head()
```

```python
df_avg_exp_age = df.groupby('age')['charges'].mean().to_frame().reset_index()
```

```python
df_avg_exp_age.head()
```

```python
sns.lineplot(x='age', y='charges', data=df_avg_exp_age, ci=None)
plt.xlabel('age')
plt.ylabel('charges')
plt.title('Avg Expense Vs Age')
plt.legend()#
plt.show()
```

```python
df_premium_region = df.groupby('region')['charges'].mean().to_frame().reset_index()
```

```python
df_premium_region.head()
```

```python
sns.barplot(x='region', y='charges', data=df_premium_region)
plt.show()
```

```python
df_premium_region_sex = df.groupby(['region', 'sex'])['charges'].mean().to_frame().reset_index()
```

```python
df_premium_region_sex
```

```python
sns.barplot(x='region', y='charges', hue='sex', data=df_premium_region_sex)
plt.show()
```

```python
sns.displot(x='charges', data=df, kind='hist')
plt.show()
```
